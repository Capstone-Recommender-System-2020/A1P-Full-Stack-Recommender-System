```python
import random
import pandas as pd
import numpy as np
buckets_data = pd.read_csv("New_Cleaned_Arts_Database 10.15.2020 (1).csv")
buckets_data.head()
```

```python
column_names = list(buckets_data.columns.values)
column_names
```

```python
dummy_data = pd.read_csv("Dummy Dataset 3.0.csv")
dummy_data.head()
```

### Generating User_ID

```python
from itertools import repeat 
U_Ids = []
for i in range(500):
    U_Ids.extend(repeat(("User_ID" + str(i+1)), 10))

print(str(U_Ids))
```

```python
user_df = pd.DataFrame({"User_ID": U_Ids})
frames = [user_df, dummy_data]
combined_dummy_data = pd.concat(frames)
combined_dummy_data.head(12)
```

### Generating Events Watch History

```python
def cleaned_event_bucket_list (columnName):
    list_of_events_in_col = buckets_data[columnName].to_list()
    
    Series_of_events_in_col = pd.Series(list_of_events_in_col)
    cleaned_Series = Series_of_events_in_col.dropna()
    cleaned_events_bucket= cleaned_Series.tolist()
    
    return cleaned_events_bucket
#this function converts all the events under a column/bucket to a list, then to a Series (to drop nan values), 
#then back to a list. It is easier to pick random value from list
```

```python
complete_user_event_history = []
genre_activity_type = []

for i in range (50):
    for i in range(10):
 
        col1 = random.randint(1,10)
        User_Event_History = []
        for i in range(col1):
            bucket1 = random.choice(column_names)
            #genre_activity_type.append(bucket1)
            picked_event = random.choice(cleaned_event_bucket_list(bucket1))
            User_Event_History.append(picked_event)
            complete_user_event_history.append(picked_event)
            genre_activity_type.append(bucket1)
        
        for i in range (10-col1):
            bucket2 = random.choice(column_names)
            #genre_activity_type.append(bucket2)
            picked_event = random.choice(cleaned_event_bucket_list(bucket2))
            User_Event_History.append(picked_event)
            complete_user_event_history.append(picked_event)
            genre_activity_type.append(bucket2)
```

```python
genre_df = pd.DataFrame({"Activity Type/Genre": genre_activity_type})
genre_df
```

```python
events_df = pd.DataFrame({"Event Watched": complete_user_event_history})
events_df
```

```python
merged_df = user_df.merge(events_df, how='outer', left_index=True, right_index=True)
```

```python
merged_df
```

### Generating User Ratings

```python
possible_ratings = ["thumbs up", "thumbs down", "neutral"]
submitted_ratings = []

for i in range (500):
    for i in range(10):
        
        ratings_list_completed = random.choices(possible_ratings)
        submitted_ratings.append(ratings_list_completed)
        
print(submitted_ratings)
```

```python
ratings_df = pd.DataFrame({"User Rating": submitted_ratings})
ratings_df
```

```python
new_merged_df = merged_df.merge(ratings_df, how='outer', left_index=True, right_index=True)
new_merged_df
```

```python
completed_merged_df = new_merged_df.merge(genre_df, how='outer', left_index=True, right_index=True)
completed_merged_df
```

```python
completed_merged_df.to_csv('Completed_User_Interaction_Dataset.csv')
```
